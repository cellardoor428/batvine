```python
# Import everything 
import pandas as p
% pylab inline
import datetime
from scipy import stats
```

```python
# Import pass_bin_temp csv
table = p.read_csv('../results/pass_bin_temp.csv')
```

```python
# Make a simplified table with fewer site variables (drop most habitat one) *still need to add bug density variables
dtable = p.DataFrame(data = table['detector_date'], index =None, columns=['detector_date'])
dtable.insert(1, 'date', table['date'])
dtable.insert(2, 'site', table['site'])
dtable.insert(3, 'TOTA', table['TOTA'])
dtable.insert(4, 'MYYU', table['MYYU'])
dtable.insert(5, 'TABR', table['TABR'])
dtable.insert(6, 'EPFU', table['EPFU'])
dtable.insert(7, 'ANPA', table['ANPA'])
dtable.insert(8, 'MYEV', table['MYEV'])
dtable.insert(9, 'MYTH', table['MYTH'])
dtable.insert(10, 'LANO', table['LANO'])
dtable.insert(11, 'LACI', table['LACI'])
dtable.insert(12, 'MYCA', table['MYCA'])
dtable.insert(13, 'LABL', table['LABL'])
dtable.insert(14, 'PAHE', table['PAHE'])
dtable.insert(15, 'MYLU', table['MYLU'])
dtable.insert(16, 'MYVO', table['MYVO'])
dtable.insert(17, 'COTO', table['COTO'])
dtable.insert(17, 'EUPE', table['EUPE'])
dtable.insert(18, 'extra', table['extra'])
dtable.insert(19, 'pair', table['pair'])
```

```python
# Get rid of the extras
# First set a condition indicating extras == 0, aka not an extra
condition = (table['extra'] == 0)
omit_extra = np.extract(condition, table['detector_date']) # then extract the detector_date meeting that criteria

# make the detector_date column (omitting the extras) a dataframe that you can then join with the dtable
dd_no_extra = p.DataFrame(data = omit_extra, index =None, columns=['detector_date'])

# Join the omitted dd_no_extra table with the dtable, using only keys from the omitted dataframe, eliminating the extras
join = p.merge(dd_no_extra, table, on ='detector_date', how ='left')
```

```python
# Now set pair vs. not pair as a condition
c1 = (join['pair'] == 0)
c2 = (join['pair'] == 1)

# Now extract the total, EPFU, MYYU, and TABR columns for both pairs and not pairs
r_tota = np.extract(c1, join['TOTA'])
p_tota = np.extract(c2, join['TOTA'])

r_myyu = np.extract(c1, join['MYYU'])
p_myyu = np.extract(c2, join['MYYU'])

r_epfu = np.extract(c1, join['EPFU'])
p_epfu = np.extract(c2, join['EPFU'])

r_tabr = np.extract(c1, join['TABR'])
p_tabr = np.extract(c2, join['TABR'])
```

```python
# Now make a figure of the mean nightly activity at paired sites vs. the vineyard's interior
# setup an empty figure
fig, ax1 = plt.subplots(1, 1)

# Set the x value range (will be 4: total, EPFU, MYYU, and TABR)
ind = np.arange(4)
width = 0.2 # set the column widths

# Set up the bars, so that for each spp. (or tota) there are 2 bars, the pair & not pair
ax1.bar(0, np.mean(p_tota), width, yerr = stats.sem(p_tota), ecolor ='k')
ax1.bar(0 + width, np.mean(r_tota), width, color = 'y', yerr = stats.sem(r_tota), ecolor ='k')

ax1.bar(1, np.mean(p_epfu), width, yerr = stats.sem(p_epfu), ecolor ='k')
ax1.bar(1 + width, np.mean(r_epfu), width, color = 'y', yerr = stats.sem(r_epfu), ecolor ='k')

ax1.bar(2, np.mean(p_myyu), width, yerr = stats.sem(p_myyu), ecolor ='k')
ax1.bar(2 + width, np.mean(r_myyu), width, color = 'y', yerr = stats.sem(r_myyu), ecolor ='k')

ax1.bar(3, np.mean(p_tabr), width, yerr = stats.sem(p_tabr), label = 'edge', ecolor ='k')
ax1.bar(3 + width, np.mean(r_tabr), width, color = 'y', yerr = stats.sem(r_tabr), label = 'interior', ecolor ='k')

# do formatting stuff
ax1.set_xticks(ind + width)
ax1.set_xticklabels(('Total', 'EPFU', 'MYYU', 'TABR'))
ax1.set_ylabel('mean nightly passes')
ax1.legend()
ax1.set_title('Activity in the vineyard interior vs. edge habitat')
fig.set_dpi(300)
fig.set_size_inches(8, 6)

# Save the figure
fig.savefig('../results/mean_nightly_passes_pair.png')
```

```python

# make a table with site and Total activity, and 3 most common species
ttable = p.DataFrame(data = join['site'], index =None, columns=['site'])
ttable.insert(1, 'TOTA', join['TOTA']) 
ttable.insert(2, 'EPFU', join['EPFU'])
ttable.insert(3, 'MYYU', join['MYYU'])
ttable.insert(4, 'TABR', join['TABR'])

print ttable.columns
```

```python
# Now I want to make a figure with the total activity by site (I need to remove the extras)
# Here I am adding a column for the total nightly activity - the activity of the top 3 spp. (which I
# will be including in the stacked bars

tota_rest = []

for i in range(len(ttable['site'])):
        tota_rest.append(ttable.values[i][1] - np.sum(ttable.values[i][2:]))
     
ttable.insert(5, 'rest', tota_rest)
```

```python
# I need to group the ttable by total nightly activity in descending order, but first (because I can't sort by the sem,
# I need to created variables, and then a dataframe with the sem's for each spp. and the rest of the spp. Also I will 
# append the site to the dataframe, so  Ican them merge this with the sorted ttable, on the key 'site'.
unique_site = np.unique(ttable['site'])
# Setup empty lists to hold the sems for each spp.
epfu_sem = []
myyu_sem = []
tabr_sem = []
rest_sem = []
site_sem = []

# loop through the totals for each spp. by site, extracting the sem
for i in unique_site:
    mini_table = ttable[ttable['site'] == i]
    epfu_sem.append(stats.sem(mini_table['EPFU']))
    myyu_sem.append(stats.sem(mini_table['MYYU']))
    tabr_sem.append(stats.sem(mini_table['TABR']))
    rest_sem.append(stats.sem(mini_table['rest']))
    site_sem.append(i)
# Create a dataframe with the unique sites and the corresponding sem's
sem = p.DataFrame(data = site_sem, index = None, columns = ['site'])
sem.insert(1, 'epfu_sem', epfu_sem)
sem.insert(2, 'myyu_sem', myyu_sem)
sem.insert(3, 'tabr_sem', tabr_sem)
sem.insert(4, 'rest_sem', rest_sem)
```

```python
# First, group the ttable by site, obtaing means for the total, each spp. and the "rest" of the spp.
sites_mean = ttable.groupby('site').mean().reset_index()

# Now merge this groups ttable with the sem table
site_means = p.merge(sites_mean, sem, on = 'site', how = 'left')

# then sort the joined site_mean table by the mean total activity in descending order
site_mean = site_means.sort(columns = 'TOTA', ascending = False).reset_index()
```

```python
# Create a stacked bar graph of the mean nightly activy of the 3 dominant spp. and the rest, by site
fig, ax1 = plt.subplots(1, 1)

# set up parameters
ind = np.arange(32)
width = 0.5 # set the column widths
site_label = site_mean['site']

# define the bars
ax1.bar(ind, site_mean['EPFU'], width, color = 'b', label = 'E. Fuscus', yerr = site_mean['epfu_sem'], ecolor = 'k')
ax1.bar(ind, site_mean['MYYU'], width, color = '0.75', bottom = site_mean['EPFU'], label = 'M. yumanensis', yerr = site_mean['myyu_sem'], ecolor = 'b')
ax1.bar(ind, site_mean['TABR'], width, color = 'k', bottom = site_mean['EPFU'] + site_mean['MYYU'], label = 'T. brasiliensis', yerr = site_mean['tabr_sem'], ecolor = 'b')
ax1.bar(ind, site_mean['rest'], width, color = 'c', bottom = site_mean['EPFU'] + site_mean['MYYU'] + site_mean['TABR'] , label = 'other spp.', yerr = site_mean['rest_sem'], ecolor = 'k')

# do formatting stuff
ax1.set_xticks(np.arange(32))
ax1.set_xlim(0, 32)
ax1.set_xticklabels(site_label, rotation = 50)
ax1.set_ylabel('mean nightly passes')
ax1.set_xlabel('site abbreviation')
ax1.legend()
ax1.set_title('Activity & species composition at vineyard sites')
fig.set_dpi(300)
fig.set_size_inches(14, 8)

# Save the figure
fig.savefig('../results/activity_spp_site.png')
```
