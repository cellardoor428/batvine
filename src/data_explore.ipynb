```python
# Make a new ttable, grouped by site to get each unique site, and the mean number of passes

# Setup empty lists to hold the mean total, EPFU, MYYU, and TABR passes 
mean_TOTA_col = []
mean_EPFU_col = []
mean_MYYU_col = []
mean_TABR_col = []

# Group ttable by site 
gtable = ttable.groupby('site').sum().reset_index()

# Loop through gtable and pull out mean counts for total, EPFU, MYYU, and TABR passes by site
for i in range(len(gtable['site'])):
    
    # append the mean nightly passes for each site 
    mean_TOTA_col.append(np.mean(gtable.values[i][1]))
    mean_EPFU_col.append(np.mean(gtable.values[i][2]))
    mean_MYYU_col.append(np.mean(gtable.values[i][3]))
    mean_TABR_col.append(np.mean(gtable.values[i][4]))

# Append these mean values to the table 
gtable['mean_TOTA'] = mean_TOTA_col
gtable['mean_EPFU'] = mean_EPFU_col
gtable['mean_MYYU'] = mean_MYYU_col
gtable['mean_TABR'] = mean_TABR_col

# Check that the columns were added 
print gtable.columns
```

```python
# Make a new column to hold the the number of nights each site was sampled
# Create a list for all the unique sites
unique_site = np.unique(table['site'])

# Setup an empty list to hold the number of times each site was sampled, and site names excluding extras
n_site = []
site_col = []


# Loop through each uniqe site and append the number of times it occurs to the n_site list
for i in unique_site:
    # Make a mini table defining the site as i
    mini_table = table[table['site'] == i]
    
     # Drop the extras
    list_site = list(i)
    split_site = list_site[4] # Pull of the last character in the site name
    
    # If the last character is X, drop site from the unique site list
    if split_site != 'X':
      
       # Append the number of times each site occurs 
       n_site.append(len(mini_table))
        
       # Append the sites, omitting the extras
       site_col.append(i)
    

# Do a quick visual check
print len(unique_site)
print len(n_site)
print site_col
print n_site

```
