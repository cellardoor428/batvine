```python
# Make a new column to hold the the number of nights each site was sampled
# Create a list for all the unique sites
unique_site = np.unique(table['site'])

# Setup an empty list to hold the number of times each site was sampled, and site names excluding extras
n_site = []
site_col = []


# Loop through each uniqe site and append the number of times it occurs to the n_site list
for i in unique_site:
    # Make a mini table defining the site as i
    mini_table = table[table['site'] == i]
    
     # Drop the extras
    list_site = list(i)
    split_site = list_site[4] # Pull of the last character in the site name
    
    # If the last character is X, drop site from the unique site list
    if split_site != 'X':
      
       # Append the number of times each site occurs 
       n_site.append(len(mini_table))
        
       # Append the sites, omitting the extras
       site_col.append(i)
    

# Do a quick visual check
print len(unique_site)
print len(n_site)
print site_col
print n_site

# Make the unique site (omitting the extras) and the number of nights it was sampled into a dataframe that you can then merge with the
# original data table
u_sites = p.DataFrame(data=site_col, index=None, columns=['site'])
u_sites.insert(1, 'n_nights', n_site)

# Make a figure of the site by sampling effort!!!
```

```python
# Now merge the full table & the u_sites table, using site as the key, so that the sampling effort will be appended to each site.
join = p.merge(table, u_sites, on ='site', how ='left') 

# Now make this joined table into a csv and visually check it over
join.to_csv('../results/both_years_full_sample_effort.csv', index =False)
```

```python
# Make some plots of the total nightly activity vs. date

# Make an empty column to hold date
dt_date = []
# and one to hold both years' dates pooled together
pooled_years =[]

# Loop through the date column and convert the dates to datetime objects
for i in table['date']:

    tdate = str(i)
    tyear = int(tdate[:4])
    tmonth = int(tdate[4:6])
    tday = int(tdate[6:])
    dt_date.append(datetime.datetime(tyear, tmonth, tday))
    pooled_years.append(datetime.datetime(2000, tmonth, tday))
        
# Make a figure of the total activity & log transformed vs. date 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2)
ax1.plot_date(dt_date, TOTA,)
ax2.plot_date(pooled_years, TOTA)
ax3.plot_date(pooled_years, np.log10(TOTA))
ax4.plot_date(pooled_years, np.log10(TOTA)) # Need something else to plot
fig.set_size_inches(15,10)       
```
