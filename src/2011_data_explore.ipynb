```python
# Loop through the date column and convert the dates to datetime objects
# Setup an empty list to hold the datetime dates
dt_date = []

# Loop through the date column, separate year, month, and day, then define as a datetime object
for i in table['date']:
    tdate = str(i)
    tyear = int(tdate[:4])
    tmonth = int(tdate[4:6])
    tday = int(tdate[6:])
    dt_date.append(datetime.datetime(tyear, tmonth, tday)) 
```

```python
# Plot the total activity 
fig, ax1 = plt.subplots(1, 1)
ax1.plot_date(dt_date, np.log10(TOTA))

ax1.set_xlabel('date')
ax1.set_ylabel('total nightly passes (log10)')

fig.set_size_inches(10,6)

fig.savefig('../results/2011_activity_date.png')
```
